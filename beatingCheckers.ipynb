```python
class Piece:
    def __init__(self, x, y, color):
        self.x = x
        self.y = y
        self.color = color
        
    def move(self, deltaX, deltaY):
        self.x += deltaX
        self.y += deltaY
        
    def draw(self, board):
        board.draw_piece(self)
        
```
